```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv("data_after_feature_engg.csv")
```

```python
df.sample(6)
```

```python
df['Text_info'] = df['Text_info'].str.lower()
```

```python
df['Text_info'][9]
```

```python
def remove_punctuation_numbers(input_text):
    
    cleaned_text = re.sub(r'[^A-Za-z\s]', '', input_text)
    return cleaned_text
```

```python
df['Text_info'] = df['Text_info'].apply(remove_punctuation_numbers)
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

def remove_stopwords(input_text):
    stop_words = set(stopwords.words('english'))
    words = re.findall(r'\b\w+\b', input_text)  # Tokenize text
    filtered_text = ' '.join([word for word in words if word.lower() not in stop_words])
    return filtered_text
```

```python
df['Text_info'] = df['Text_info'].apply(remove_stopwords)
```

```python
df['Text_info'][9]
```

```python
from nltk.stem import SnowballStemmer

nltk.download('punkt')

snowball_stemmer = SnowballStemmer("english")

def stem_text(text):
    # Split text into words, stem each word, and rejoin
    stemmed_text = ' '.join([snowball_stemmer.stem(word) for word in text.split()])
    return stemmed_text
```

```python
df['Text_info'] = df['Text_info'].apply(stem_text)
```

```python
import inflect


p = inflect.engine()

def plural_to_singular(text):
   
    singular_text = ' '.join([p.singular_noun(word) if p.singular_noun(word) else word for word in text.split()])
    return singular_text
```

```python
df['Text_info'] = df['Text_info'].apply(plural_to_singular)
```

```python
df['Text_info'][9]
```

```python
def remove_repeated_words(text):
    words = text.split()
    unique_words = []
    
    for word in words:
        if word not in unique_words:
            unique_words.append(word)
    
    return ' '.join(unique_words)
```

```python
df['Text_info'] = df['Text_info'].apply(remove_repeated_words)
```

```python
df['Text_info'][789]
```

```python
def remove_words(input_text):
    words_to_remove = [
        "facing", "bedroom", "bathroom", "balconi", "corner", "properti",
        "furnishing", "price", "crore", "per", "sqft", "converted", "area",
        "floor", "carpet", "super", "overlooking", "main", "flat", "age",
        "sector", "num", "luxury", "score", "bhk", "gurgaon"
    ]
    
    
    filtered_words = [word for word in input_text.split() if word.lower() not in words_to_remove]
    
   
    return ' '.join(filtered_words)
```

```python
df['Text_info'] = df['Text_info'].apply(remove_words)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


text_data = df['Text_info'] 

# Initialize TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=30)  # Adjust max_features as needed

# Fit and transform the text data
tfidf_vectors = tfidf_vectorizer.fit_transform(text_data)

# Convert to DataFrame if needed
tfidf_df = pd.DataFrame(tfidf_vectors.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# Optional: Concatenate with original DataFrame
df1 = pd.concat([df.reset_index(drop=True), tfidf_df], axis=1)

# View the resulting DataFrame with vectorized features
df1.head()
```
