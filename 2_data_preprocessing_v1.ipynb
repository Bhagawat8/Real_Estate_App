```python
def price_Per_Sqft(per_sqft):
   
    if isinstance(per_sqft, str):
       
        if '@' in per_sqft:
            return per_sqft.replace('@ ', '').replace(' per', '')
            
    return per_sqft 

```

```python
df['price_Per_Sqft'] = df['Per Sqft'].apply(price_Per_Sqft)
```

```python
df['price_Per_Sqft'].value_counts()
```

```python
import re
def extract_number_and_unit(value):

    if isinstance(value, str):
        match = re.match(r'([\d,]+)\s*(\w+\.\w+|\w+)', value)
        if match:
            number = match.group(1).replace(",", "") 
            unit = match.group(2)  
            return pd.Series([number, unit])
    return pd.Series([None, None])

```

```python
# Apply the function to the 'price_Per_Sqft' column
df[['number', 'unit']] = df['price_Per_Sqft'].apply(extract_number_and_unit)
```

```python
def extract_number_and_convert(value):
    if isinstance(value, str):
        # Extract numeric part (removing commas) and unit part using regex
        match = re.match(r'([\d,]+)\s*(\w+\.\w+|\w+)', value)
        if match:
            number = float(match.group(1).replace(",", ""))  # Remove commas and convert to float
            
            # Conversion to sq.ft.
            if "sq.m." in value:
                number = number * 10.764
                return number
            elif "sq.yards" in value:
                number = number * 9
                return number
            elif "sq.ft." in value:                         
                return number
    return None



```

```python
# Apply the function to convert all areas to sq.ft.
df['price_Per_Sqft_converted'] = df['price_Per_Sqft'].apply(extract_number_and_convert)
```

```python
def extract_floors(value):
    try:

        patterns = [
            r'(\d+)[a-z]{2}\s+of\s+(\d+)',  
            r'(\d+)\s+of\s+(\d+)',         
            r'(Ground|Basement|Lower Ground|Upper Ground)\s+of\s+(\d+)' 
        ]
        
        for pattern in patterns:
            match = re.search(pattern, value, re.IGNORECASE)
            if match:
                floor_number = match.group(1)
                
                # Handle specific cases
                if floor_number.lower() == "ground":
                    floor_number = 0
                elif floor_number.lower() == "lower ground":
                    floor_number = -1
                elif floor_number.lower() == "basement":
                    floor_number = -2
                elif floor_number.isdigit():
                    floor_number = int(floor_number)
                
                total_floors = int(match.group(2))
                
                return floor_number, total_floors

        return value, value
    except Exception:

        return value, value

df[['floor_number', 'total_floors']] = df['Floor Number'].apply(lambda x: pd.Series(extract_floors(x)))
```
