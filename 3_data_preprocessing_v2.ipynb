```python
import re
def extract_sector_number(sector_name):
    match = re.search(r'\bsector[-\s]?(?P<number>\d+)', sector_name.lower())
    if match:
        return f"sector {match.group('number')}"
    return None
```

```python
df['Extracted Sector'] = df['sector'].apply(extract_sector_number)
```

```python
def categorize_age_possession(value):
    if pd.isna(value):
        return "Undefined"
    if "0 to 1 Year Old" in value or "Within 6 months" in value or "Within 3 months" in value:
        return "New Property"
    if "1 to 5 Year Old" in value:
        return "Relatively New"
    if "5 to 10 Year Old" in value:
        return "Moderately Old"
    if "10+ Year Old" in value:
        return "Old Property"
    if "Under Construction" in value or "By" in value:
        return "Under Construction"
    try:
        # For entries like 'May 2024'
        int(value.split(" ")[-1])
        return "Under Construction"
    except:
        return "Undefined"
```

```python
df['flat_age'] = df['Age'].apply(categorize_age_possession)
```
