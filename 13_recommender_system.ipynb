```python
unique_features = set(feature for feature_list in df['Features'] for feature in feature_list)
unique_features = sorted(unique_features)  
```

```python
feature_to_index = {feature: i for i, feature in enumerate(unique_features)}

def vectorize_features(features_list):
    vector = [0] * len(unique_features)
    for feature in features_list:
        vector[feature_to_index[feature]] = 1
    return vector

df['Features_vector'] = df['Features'].apply(vectorize_features)
```

```python
location_to_index = {location: i for i, location in enumerate(unique_locations)}

def vectorize_locations(locations_list):
    vector = [0] * len(unique_locations)
    for location in locations_list:
        vector[location_to_index[location]] = 1
    return vector

df['locations_vector'] = df['Nearby Locations'].apply(vectorize_locations)
```

```python
unique_rooms = set(room for rooms_list in df['additional_room'] for room in rooms_list)
unique_rooms  = sorted(unique_rooms )  
```

```python
room_to_index = {room: i for i, room in enumerate(unique_rooms)}

def vectorize_rooms(rooms_list):
    vector = [0] * len(unique_rooms)
    for room in rooms_list:
        vector[room_to_index[room]] = 1
    return vector

df['rooms_vector'] = df['additional_room'].apply(vectorize_rooms)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_properties_strict(no_bedrooms, min_price, max_price, df, cosine_sim):
    filtered_df = df[(df['No_Bedroom'] == no_bedrooms) & (df['Price_in_Crore'] >= min_price) & (df['Price_in_Crore'] <= max_price)]
    
    if filtered_df.empty:
        return pd.DataFrame({'Message': ['No properties found matching the criteria']})

    filtered_indices = filtered_df.index.tolist()
    filtered_cosine_sim = cosine_sim[np.ix_(filtered_indices, filtered_indices)]

    query_idx = 0
    sim_scores = list(enumerate(filtered_cosine_sim[query_idx]))

    sim_scores = [(idx, score) for idx, score in sim_scores if 0 < score < 1] 
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)#[:5]  

    if not sim_scores:
        return pd.DataFrame({'Message': ['No similar properties found with score < 1']})

    similar_property_indices = [filtered_indices[i[0]] for i in sim_scores]

    recommendations_df = df.iloc[similar_property_indices].copy()
    recommendations_df['SimilarityScore'] = [score[1] for score in sim_scores]
    recommendations_df = recommendations_df.sort_values(by=['SimilarityScore'] , ascending = False)[:5]
    recommendations_df = recommendations_df.drop(columns = ['SimilarityScore'])
    

    return recommendations_df

recommended_properties1 = recommend_properties_strict(
    no_bedrooms=5,
    min_price=8,
    max_price=15,
    df=df_or, 
    cosine_sim=cosine_sim  
)

recommended_properties1
```
